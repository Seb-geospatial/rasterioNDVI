# rasterioNDVI Demonstration Notebook
This Jupyter notebook is for demonstrating the usage and output of the `rasterioNDVI.py` script.

First, we will import `rasterio` and `matplotlib` for use in plotting the input and output data for demonstration purposes.

````python
# Import packages
import rasterio
from rasterio.plot import show
from matplotlib import pyplot
````

Next, lets use `rasterio` to examine the NIR-band and red-band input rasters being used in this demonstration. These rasters are centered on an area just west of Montreal and were collected by Landsat 8 courtesy of [USGS Earth Explorer](https://earthexplorer.usgs.gov/).

````python
# Use rasterio to open raster images for viewing
NIR_raster = rasterio.open('./data/NIR_band_1.tif')
red_raster = rasterio.open('./data/red_band_1.tif')
````

````python
# Used matplotlib to plot both rasters in one chart
fig, (axnir, axr) = pyplot.subplots(1,2)
show((NIR_raster), ax = axnir, title = 'Near-Infrared Band Raster')
show((red_raster), ax = axr, title = 'Red Band Raster')
pyplot.show()
````

Now we can see the data we are working with. It is important to note that the previous steps are not necessary and are only employed for demonstration purposes. Running the `rasterioNDVI.py` script and specifying the input and output paths as command-line arguments is all that is required to generate a NDVI raster image from red and NIR band input rasters.

To use `rasterioNDVI.py`, simply run the python script and specify the path to the NIR-band image, the red-band image, and the output image path as command-line arguments as seen below. After this, rasterio will be used to show the resulting NDVI raster.

````python
! python ../rasterioNDVI.py ./data/NIR_band_1.tif ./data/red_band_1.tif ./data/ndvi_1.tif

NDVI_raster = rasterio.open('./data/ndvi_1.tif')
show(NDVI_raster, title = 'NDVI Raster')
````

Additionally, the subprocess package can be used with the subprocess.run() command to run the above terminal command from within other python scripts (assuming the directory of the rasterioNDVI.py script is specified). This may be useful if you wish to automate the use of this script for batch processes.

