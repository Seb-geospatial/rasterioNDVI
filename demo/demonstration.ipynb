# rasterioNDVI Demonstration Notebook
This Jupyter notebook is for demonstrating the usage and output of the `rasterioNDVI.py` module.

First, we will import the module, as well as `rasterio` and `matplotlib` for use in plotting the input and output data.

```python
# Import packages
import rasterio
from rasterio.plot import show
from matplotlib import pyplot

#import rasterioNDVI.py
```

Next, lets use `rasterio` to examine the NIR-band and red-band input rasters being used in this demonstration. These rasters are centered on an area just north-west of Montreal and were collected by Landsat 8 courtesy of [USGS Earth Explorer](https://earthexplorer.usgs.gov/).

```python
# Use rasterio to open raster images for viewing
NIR_raster = rasterio.open('./data/NIR.tif')
red_raster = rasterio.open('./data/red.tif')
```

```python
# Used matplotlib to plot both rasters in one chart
fig, (axnir, axr) = pyplot.subplots(1,2)
show((NIR_raster), ax = axnir, title = 'Near-Infrared Band Raster')
show((red_raster), ax = axr, title = 'Red Band Raster')
pyplot.show()
```

Now we can see the data we are working with. It is important to note that the previous steps are not necessary when using `rasterioNDVI` and are only employed for demonstration purposes. The following function call of `rasterioNDVI()` is all that is required to generate a NDVI raster image from red and NIR banded input rasters.
